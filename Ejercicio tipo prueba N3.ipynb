```python
import random

datos_personas=[]

def validar_rut(rut):
  rut = rut.replace(".", "").replace("-", "")
  if len(rut) != 9 or not rut[:-1].isdigit():
    return False
  multiplicadores = [2, 3, 4, 5, 6, 7, 2, 3]
  total = sum(int(digito) * multiplicador for digito, multiplicador in zip(rut[:-1][::-1], multiplicadores))
  resto = total % 11
  digito_verificador = str(11 - resto)
  if digito_verificador == "11":
    digito_verificador = "0"
  elif digito_verificador == "10":
    digito_verificador = "K"
  return rut[-1].upper() == digito_verificador


#INGRESAR Y GRABAR DATOS SOLICITADOS
def grabar():
  rut=input('Ingresa su rut: ')
  print("===========================================================================")
  while not validar_rut(rut):
    print('ERROR: El rut ingresado es erroneo o no esta registrado.')
    print("===========================================================================")
    rut=input('Ingresa su rut: ')
    print("===========================================================================")

  nombre=input('Ingrese su nombre completo: ')
  print("===========================================================================")
  while len(nombre)<8:
      print('NOMBRE INVALIDO, Ingrese al menos 8 caracteres.')
      nombre=input('Ingrese su nombre completo: ')
      print("===========================================================================")


  edad=int(input('Ingrese su edad: '))
  print("===========================================================================")
  while edad<=0 or edad>=120:
      print('Ingrese su edad real. (entre 0 y 120)')
      print("===========================================================================")
      edad=int(input('\nIngrese su edad: '))
      print("===========================================================================")

  fech_nac = None

  while fech_nac is None or not fech_nac:
      fech_nac = input('Ingrese su fecha de nacimiento (DD/MM/AAAA): ')
      print("===========================================================================")

      if not fech_nac:
          print("Debe ingresar una fecha. Intente nuevamente.")

  print("===========================================================================")

  while True:
      partes = fech_nac.split("/")
      if len(partes) != 3 or not all(p.isdigit() for p in partes):
          print("===========================================================================")
          print("Formato de fecha incorrecto. Intente nuevamente.")
          print("===========================================================================")
          fech_nac = input('Ingrese su fecha de nacimiento (DD/MM/AAAA): ')

      else:
          dia, mes, año = map(int, partes)
          if 1 <= dia <= 31 and 1 <= mes <= 12 and 1900 <= año <= 2023:
              break
          else:
              print('Fecha Inválida')
              fech_nac = input('Ingrese su fecha de nacimiento (DD/MM/AAAA): ')

  estado_civil=input('Ingrese su estado civil (Soltero/a - Casado/a - Divorciado/a - Viudo/a): ').lower()
  print("===========================================================================")
  while not estado_civil=='soltero' and estado_civil=='soltera' and estado_civil=='soltero/a' and estado_civil=='casado' and estado_civil=='casada' and estado_civil=='casado/a' and estado_civil=='divorciado' and estado_civil=='divorciada' and estado_civil=='divorciado/a' and estado_civil=='viudo' and estado_civil=='viuda' and estado_civil=='viudo/a':
      print('Estado civil no valido. Ingrese un estado civil valido\n(Soltero/a - Viudo/a - Separado/a - Divorciado/a)')
      print("===========================================================================")
      estado_civil=input('Ingrese su estado civil (Soltero/a - Casado/a - Divorciado/a - Viudo/a): ').lower()
      print("===========================================================================")

  pais=input('Ingrese su pais de origen: ').lower()
  print("===========================================================================")
  while not pais.isalpha():
      print('EL NOMBRE DEL PAIS DEBE CONTENER SOLO LETRAS.')
      print("===========================================================================")
      pais=input('\nIngrese su pais de origen: ').lower()
      print("===========================================================================")
  ciudad=input('Ingrese su ciudad de origen: ').lower()
  print("===========================================================================")
  while not ciudad.isalpha():
      print('EL NOMBRE DE LA CIUDAD DEBE CONTENER SOLO LETRAS.')
      print("===========================================================================")
      ciudad=input('\nIngrese su ciudad de origen: ').lower()
      print("===========================================================================")

  persona={
      'rut': rut,
      'nombre': nombre,
      'edad': edad,
      'fecha_de_nacimiento': {
          'dia':dia,'mes':mes,'año':año
      },
      'pais_nacimiento':pais,
      'ciudad_nacimiento':ciudad,
      'estado_civil':estado_civil
  }
  datos_personas.append(persona)
  print('\nDatos personales guardados')

#BUSCAR DATOS INGRESADOS CON ANTERIORIDAD
def buscar():
    rut=input('Ingresa su rut: ')
    print("===========================================================================")
    while not validar_rut(rut):
      print('ERROR: El rut ingresado es erroneo o no esta registrado.')
      print("===========================================================================")
      rut=input('Ingresa su rut: ')
      print("===========================================================================")
    for persona in datos_personas:
      if persona['rut']==rut:
          print('')
          print('Información de la persona:')
          print('Nombre: ', persona['nombre'])
          print('RUT: ', persona['rut'])
          print('Fecha de nacimiento: {}/{}/{}'.format(
              persona['fecha_de_nacimiento']['dia'],
              persona['fecha_de_nacimiento']['mes'],
              persona['fecha_de_nacimiento']['año']))
          print('Edad: ', persona['edad'])
          print('Estado civil: ', persona['estado_civil'])
          print('Pais de nacimiento:', persona['pais_nacimiento'])
          print('Ciudad natal:', persona['ciudad_nacimiento'])
          if persona['pais_nacimiento']=='chile':
              print('Ciudadano con nacionalidad Chilena.')
          else:
              print('Ciudadano no cuenta con nacionalidad Chilena')
      else:
        print('PERSONA NO REGISTRADA EN NUESTRA BASE DE DATOS O RUT INVALIDO')

#IMPRIMIR CERTIFICADOS SOLICITADOS POR EL CLIENTE
def imprimir_cert():
  print("===========================================================================")
  rut=input('Ingresa su rut: ')
  print("===========================================================================")
  while not validar_rut(rut):
    print('ERROR: El rut ingresado es erroneo o no esta registrado.')
    print("===========================================================================")
    rut=input('Ingresa su rut: ')
    print("===========================================================================")

  persona_encontradaf = False

  for persona in datos_personas:
    if persona["rut"] == rut:
        persona_encontrada = persona
        break

  if persona_encontrada:
      print("Certificados de la persona con RUT:", persona_encontrada["rut"])
      print("Nombre:", persona_encontrada["nombre"])
      persona_encontradaf = True

  if persona_encontradaf:
    while True:
      print("\nSubmenú de Certificados:")
      print("1. Certificado de Nacimiento")
      print("2. Certificado de Estado Civil")
      print("3. Certificado de Pertenencia")
      print("4. Volver al Menú Principal")

      print("===========================================================================")
      opcion2 = int(input("Seleccione una opción: "))
      print("===========================================================================")
      if opcion2==1:
        print('')
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("          CERTIFICADO DE NACIMIENTO         ")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("CIRCUNSCRIPCIÓN:\t",persona["ciudad_nacimiento"].upper())
        print("AÑO DE INSCRIPCIÓN:\t",persona['fecha_de_nacimiento']['año'])
        print("NOMBRE:\t\t\t",persona['nombre'].upper())
        print("RUN:\t\t\t",persona["rut"])
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print('')

      elif opcion2==2:
        print('')
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("          CERTIFICADO DE ESTADO CIVIL         ")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("NOMBRE:\t\t\t",persona["nombre"].upper())
        print("RUN:\t\t\t",persona["rut"])
        print("ESTADO CIVIL:\t\t",persona["estado_civil"].upper())
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print('')

      elif opcion2==3:
        print('')
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("          CERTIFICADO DE PERTENENCIA         ")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("NOMBRE:\t\t\t",persona["nombre"].upper())
        print("RUN:\t\t\t",rut)
        print("CIUDAD DE NACIMIENTO:\t",persona["ciudad_nacimiento"].upper())
        print("PAIS DE NACIMIENTO:\t", persona["pais_nacimiento"].upper())
        print("")
        print("SE ACREDITA QUE LA PERSONA")
        print("")
        print("NOMBRE:\t\t\t",persona["nombre"].upper(),)
        print("RUN:\t\t\t",persona["rut"])
        print("PERTENECE A:\t\t",persona["pais_nacimiento"].upper())
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print('')

      elif opcion2==4:
        print('')
        print('Volviendo al Menú principal.')
        print('')
        break

      else:
        print('OPCION IVALIDA')

#ELIMINAR REGISTRO CON RUT DE CLIENTE
def eliminar():
    rut=input('Ingresa el rut: ')
    print("===========================================================================")
    while not validar_rut(rut):
      print('ERROR: El rut ingresado es erroneo o no esta registrado.')
      print("===========================================================================")
      rut=input('Ingresa el rut: ')
      print("===========================================================================")

    for i, persona in enumerate(datos_personas):
        if persona["rut"] == rut:
            print('Persona eliminada con exito de los registros.')
            print("===========================================================================")
            persona_eliminada = datos_personas.pop(i)
            break

#MOSTRAR LA CANTIDAD DE PERSONAS REGISTRADAS CON SU INFORMACION
def mostrar():
  cant_personas=0
  cant_personas=len(datos_personas)
  print(f'Personas registradas: {cant_personas}')
  for persona in datos_personas:
      print('')
      print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
      print('Información de la persona:')
      print('Nombre: ', persona['nombre'])
      print('RUT: ', persona['rut'])
      print('Fecha de nacimiento: {}/{}/{}'.format(
          persona['fecha_de_nacimiento']['dia'],
          persona['fecha_de_nacimiento']['mes'],
          persona['fecha_de_nacimiento']['año']))
      print('Edad: ', persona['edad'])
      print('Pais de nacimiento:', persona['pais_nacimiento'])
      print('Ciudad natal:', persona['ciudad_nacimiento'])
      print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
      print('')


def imprimir_menu():
  print("===========================================================================")
  print('')
  print("------ MENÚ ------")
  print("1. Grabar")
  print("2. Buscar")
  print("3. Imprimir certificados")
  print("4. Eliminar")
  print("5. Mostrar personas guardadas")
  print("6. Salir")
  print('')

def main():
  print("Bienvenido al programa Registro Civil Totem Autoservicio")
  print("Versión 2.0\n")

if __name__ == "__main__":
  main()
  while True:
    imprimir_menu()
    print("===========================================================================")
    opcion = input("Ingrese el número de la opción que desea seleccionar: ")
    print("===========================================================================")
    if opcion == "1":
      grabar()
    elif opcion == "2":
      buscar()
    elif opcion == "3":
      imprimir_cert()
    elif opcion == "4":
      eliminar()
    elif opcion == "5":
      mostrar()
    elif opcion == "6":
      print("Gracias por utilizar el Totem de Autoservicio de Registro Civil. Hasta luego.")
      print("===========================================================================")
      print('')
      print('')
      break
    else:
      print("OPCION IVALIDA. Ingrese una opción válida.")
```
